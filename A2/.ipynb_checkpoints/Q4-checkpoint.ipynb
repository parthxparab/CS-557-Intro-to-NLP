## Unigram model for text1 corpus

```python
unigram = {}          #This dictionary stores the probability of each word
corpusLen = len(text1)
for word in text1 :
  if word not in unigram :
    print(word, text1.count(word), corpusLen)
    wordCount = text1.count(word)
    unigram[word] = wordCount/corpusLen
    
  
```

```python
#Print first five words and its probability (Contains special characters)
count = 5
for key in unigram.keys():
  if count == 0 :
    break
  count = count - 1
  print("Word : ", key, "\n Probability : ",unigram[key])
```

## Calculating perplexity

```python
def findPerplexity(testSentence) :
  perplexity = 1
  N = 0

  for word in testSentence :
    if word in unigram :
      N = N + 1
      perplexity = perplexity * (1/unigram[word])
  perplexity = pow(perplexity, 1/float(N))
  return perplexity
```

## Lexical Diversity

```python
lexical_diversity = len(set(text1))/len(text1)
print(lexical_diversity)
```

Lexical diversity helps understand how diverse a corpus is with respect to its words. The maximum possible value of lexical diversity can be one where each word only occurs once.

Perplexity on the other is used to determine how good a prediction model is. In the example above, the model used in unigram model. It is tested against two test sentences. Larger value of perplexity indicates lower probabilities of predicting next word correctly. 

One of the reasons for perplexity to be higher (ie low probability to predict next word correctly) is presence of words not present in the corpus that we traine out model on. For example, there might be words in test sentences for which there exists no unigram probability due to absence of that word in the first corpus. Thus if lexical diversity of a corpus is closer to one, we can expect that corpus to have a large number of unique words (given that the corpus size is also large). Thus one potential relationship that can be deduced is that lexical diversity close to one might result in lower perplexity and thus a good prediction model.
